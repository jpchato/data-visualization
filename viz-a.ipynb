```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
global_temps = pd.read_csv("GlobalTemperatures.csv")
global_temps
```

```python
global_temps.head(5)
```

```python
X = global_temps["LandAverageTemperature"].values
y = global_temps["LandAndOceanAverageTemperature"].values
X = X.reshape(-1,1)
# pd.DataFrame(X).fillna()
# global_temps.fillna(-99999, inplace=True)
# # https://stackoverflow.com/questions/31323499/sklearn-error-valueerror-input-contains-nan-infinity-or-a-value-too-large-for
# # Solution to error ValueError: Input contains NaN, infinity or v alue too large for dtype('float64')
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='red')
plt.xlabel("Land Average Temperature")
plt.ylabel("Land and Ocean Average Temperature")
plt.title("Land Average Temperature vs. Land and Ocean Average Temperature")
# plt.show()
plt.plot()
```

* As the average land temperature increases, so does the land&ocean temperature average
    * This seems intuitive. It would have been strange to see an inverse relationship.
