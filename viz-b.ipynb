```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df_vet = pd.read_csv("2011.csv")
df_vet.head(5)
```

```python
X = df_vet["vet_suicides"].values
y = df_vet["vet_pop"].values
X = X.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
%matplotlib inline
plt.scatter(x_train, y_train, color='red')
plt.xlabel("2011 Veteran Suicides")
plt.ylabel("2011 Veteran Population")
plt.title("Veteran Suicides vs Veteran Population")
# plt.show()
plt.plot()
```

* Veteran suicides increase as the population of veterans increases
    * Seems like an obvious observation and the data supports it
